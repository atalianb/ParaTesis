```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

$\Phi _{M} (R,z) = -\frac{GM}{\sqrt{R^{2} + (a + \sqrt{z^2 + b^2})^{2}}}$

$\rho _{M} (R,z) = \frac{b^{2}M}{4\pi} \frac{aR^{2} + (a + 3\sqrt{z^{2} + b^{2}})(a+ \sqrt{z^{2} + b^{2}})^{2}}{(R^{2} + (a + \sqrt{z^2 + b^2})^{2})^{5/2}(z^{2} + b^{2})^{3/2}}$

Con  densidades iguales y con $b/a = 0.2$, $b/a = 1.$ y $b/a = 5.$ respectivamente 

```python
R = np.linspace(-10.,10.,100)
z = np.linspace(-10.,10.,100)
```

```python
G = 4.302e-3 #pcSM(km/s)**2
M = 0.1
```

```python
def potencial(R,z,a,b):
    d = np.sqrt(R**2 + (a + np.sqrt(z**2 + b**2))**2)
    phi = -G*M/d
    return phi
```

```python
plt.plot(R, potencial(R,z,a=1,b=0.2),label=r'$a = 1, b = 0.2$')
plt.plot(R, potencial(R,z,a=1.,b=1.),label=r'$a = 1, b = 1$')
plt.plot(R, potencial(R,z,a=1.,b=5.),label=r'$a = 1, b = 5$')
plt.xlabel(r'$R$')
plt.ylabel(r'$\Phi (R)$')
plt.legend(loc='lower left', prop={'size':10})
plt.title(' z = R')
plt.show()
```

```python
def densidad(R,z,a,b):
    K = (M*b**2.)/(4.*np.pi)
    n = a*R**2. + (a + 3.*np.sqrt(z**2. + b**2.))*(a + np.sqrt(z**2. + b**2.))**2.
    d = (R**2. + (a + np.sqrt(z**2. + b**2.))**2.)**(5./2.) * (z**2. + b**2.)**(3./2.)
    rho = K*(n/d)
    return rho
```

```python
plt.plot(R, densidad(R,z,a=1,b=0.2),label=r'$a = 1, b = 0.2$')
plt.plot(R, densidad(R,z,a=1.,b=1.),label=r'$a = 1, b = 1$')
plt.plot(R, densidad(R,z,a=1.,b=5.),label=r'$a = 1, b = 5$')
plt.xlabel(r'$R$')
plt.ylabel(r'$\rho (R)$')
plt.legend(loc='upper left', prop={'size':10})
plt.title(' z = R')
plt.show()
```

```python
plt.plot(R/5., densidad(R,z,a=5.,b=1.),label=r'$a = 5, b = 1$')
plt.plot(R/1., densidad(R,z,a=1.,b=1.),label=r'$a = 1, b = 1$')
plt.plot(R/0.2, densidad(R,z,a=0.2,b=1.),label=r'$a = 0.2, b = 1$')
plt.xlabel(r'$R$')
plt.ylabel(r'$\rho (R)$')
plt.legend(loc='upper left', prop={'size':10})
plt.title(' z=R')
plt.show()
```

```python
R1 = np.linspace(-10.,10.,100)
z1 = np.linspace(0.,10.,100)
plt.plot(R1, densidad(R1,z1,a=5,b=1),label=r'$a = 5, b = 1$')
plt.plot(R1, densidad(R1,z1,a=1.,b=1.),label=r'$a = 1, b = 1$')
plt.plot(R1, densidad(R1,z1,a=1.,b=0.2),label=r'$a = 0.2, b = 1$')
plt.xlabel(r'$R$')
plt.ylabel(r'$\rho (R)$')
plt.legend(loc='upper right', prop={'size':10})
plt.show()
```
