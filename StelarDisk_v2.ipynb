```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

$$\begin{equation}
v^{2}_{c} = 2\pi G\Sigma_{d}R_{d} \left( -e^{-R/R_{d}}(1 + R_{d}/R) + R_{d}/R \right)
\end{equation}$$

$$\begin{equation}
\frac{v^{2}_{c}}{2\pi G\Sigma_{d}R_{d}} = -e^{-R/R_{d}}(1 + R_{d}/R) + R_{d}/R
\end{equation}$$

$$\begin{equation}
\frac{v^{2}_{c}}{2\pi G\Sigma_{d}R_{d}} =  R_{d} - e^{-R/R_{d}}(R_{d} + R) 
\end{equation}$$

```python
def V2_c(R,R_d=2.0):
    result = (R_d/R) - np.exp(-R/R_d) - (np.exp(-R/R_d)*(R_d/R))
    return result
```

```python
#Constants
G = 4.302e-6#(kpc/SM)*(km/s)^2
R_d_I = 2.#kpc
R_d_II = 3.2#kpc
Sigma_d_I = (1905.0e6)*0.75#SM/kpc^2
Sigma_d_II = (536.0e6)*0.75#SM/kpc^2
C_I = 2.*np.pi*G*Sigma_d_I*R_d_I
C_II = 2.*np.pi*G*Sigma_d_II*R_d_II
print C_I,C_II
```

```python
R = np.linspace(0.1,15.,100)
```

```python
plt.plot(R,V2_c(R)*C_I,label='model I')
plt.plot(R,C_II*V2_c(R,R_d=R_d_II),label='model II')
#xlim(0.,15.)
plt.xlabel(r'$R$')
plt.ylabel(r'$V^{2}_{c}/km s^{-2}$')
plt.legend(loc='upper right', prop={'size':10})
plt.show()
```

```python
plt.plot(R,np.sqrt(C_I*V2_c(R)),label='model I')
plt.plot(R,np.sqrt(C_II*V2_c(R,R_d=R_d_II)),label='model II')
#xlim(0.,15.)
plt.xlabel(r'$R$')
plt.ylabel(r'$V_{c}/km s^{-1}$')
plt.legend(loc='upper right', prop={'size':10})
plt.show()
```
